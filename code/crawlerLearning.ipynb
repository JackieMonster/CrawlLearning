```python
####实例三：非文本文件爬取实例，以图片为例（key：二进制读写文件）
import os
url = 'https://seopic.699pic.com/photo/40100/6015.jpg_wh1200.jpg'
root = "D://pics"
path = root+'//'+url.split("/")[-1]
try:
    if not os.path.exists(root):
        os.mkdir(root)#####文件夹不存在，则我们人工创建文件夹
    if not os.path.exists(path):  ###文件不存在，就写入文件
        r = requests.get(url)
        with open(path,"wb") as f:
            f.write(r.content)  #####r.content：response对象的二进制内容
            f.close()
            print("success")
    else:
        print("File exists")
except:
    print("ERROR")
```

```python
#####实例一：中国大学排名爬虫实例####
import requests 
from bs4 import BeautifulSoup
import bs4
import pandas as pd
def getHTMLtext(url): ####通用结构
    try:
        r = requests.get(url)

        r.raise_for_status()
        r.encoding = r.apparent_encoding
        soup = BeautifulSoup(r.text,'html.parser')
        return soup
    except:
        return "ERROR"


def DataDealing(rawtext):
    Datals=[]
    Inittag = rawtext.find("tbody")
    # print(Inittag.prettify())
    for tag in Inittag.children:
        if isinstance(tag,bs4.element.Tag):
            # print(tag)
            tds = tag("td")
            xiaoming = tds[1].a.string
            wangzhi = tds[1].a.attrs['href']
            dizhi = list(tds[2].strings)[0].replace("\n","")
            dizhi = dizhi.replace(" ","")
            leixin = list(tds[3].strings)[0].replace('\n','')
            leixin = leixin.replace(" ","")
            fenshu = list(tds[4].strings)[0].replace('\n','')
            fenshu = fenshu.replace(" ","")
            Datals.append([xiaoming,wangzhi,dizhi,leixin,fenshu])
    df = pd.DataFrame(Datals)    
    return df



def main():
    url = 'https://www.shanghairanking.cn/rankings/bcur/202010'
    rawtext = getHTMLtext(url)
    # print(rawtext)
    goodtext = DataDealing(rawtext)
    print(goodtext)
main()

####总结：这个实例走的是requests-bs4路线，未用到正则表达式
```
